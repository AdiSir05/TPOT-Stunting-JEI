```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from heatmap_maker import make_confusion_matrix
```

```python
kenya = pd.read_csv("kenya_corr.csv")
kenya = kenya.iloc[:,0:8]
bangladesh = pd.read_csv("bangladesh_corr.csv").iloc[:,0:8]
```

```python
fig = plt.figure(figsize = (40,20))
ax1 = fig.add_subplot(1, 2, 1) # row, column, position
ax2 = fig.add_subplot(1, 2, 2)
sns.set(font_scale = 1.85)
k = sns.heatmap(kenya.corr(), vmin=-1, vmax=1, annot=True, cmap="PuOr", ax=ax1, square=True, cbar_kws={'shrink': .3})
k.set_title('Kenya Features Correlation Heatmap', fontdict={'fontsize':35}, pad=12)
b = sns.heatmap(bangladesh.corr(), vmin=-1, vmax=1, annot=True, cmap="PuOr", ax=ax2, square=True, cbar_kws={'shrink': .3})
b.set_title('Bangladesh Features Correlation Heatmap', fontdict={'fontsize':35}, pad=12)
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.65, 
                    wspace=0.35, 
                    hspace=0.2)
plt.savefig('corr_heatmaps.png', dpi=600, bbox_inches='tight')
```

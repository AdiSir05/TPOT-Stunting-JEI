```python
import sys
# !{sys.executable} -m pip install seaborn
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from heatmap_maker import make_confusion_matrix
```

```python
kenya = pd.read_csv("kenya_scores.csv")
bangladesh = pd.read_csv("bangladesh_scores.csv")
```

```python
cf1 = np.array([kenya.iloc[0, 8:10], kenya.iloc[0, 10:12]])
cf1
```

```python
cf_list = []
for i in range(11):
    cf_list.append(np.array([kenya.iloc[i, 8:10], kenya.iloc[i, 10:12]]).astype(int))
```

```python
kenya_title = ["Kenya Soft Voting", "Pipeline 100", "Pipeline 12", "Pipeline 124", "Pipeline 24", 
         "Pipeline 34", "Pipeline 44", "Pipeline 50", "Pipeline 68", "Pipeline 72", "Pipeline 75"]
bangladesh_title = ["Bangladesh Soft Voting", "Pipeline 100", "Pipeline 12", "Pipeline 124", "Pipeline 24", 
         "Pipeline 34", "Pipeline 44", "Pipeline 50", "Pipeline 68", "Pipeline 72", "Pipeline 75"]
labels = ["True Neg","False Pos","False Neg","True Pos"]
categories = ["No", "Yes"]
sns.set(font_scale = 1.75)
```

```python
fig = plt.figure(figsize = (20,20)) # width x height
fig.suptitle('Confusion Matrices for Kenya Models', fontsize=30, y=0.7)
ax1 = fig.add_subplot(3, 4, 1) # row, column, position
ax2 = fig.add_subplot(3, 4, 2)
ax3 = fig.add_subplot(3, 4, 3)
ax4 = fig.add_subplot(3, 4, 4)
ax5 = fig.add_subplot(3, 4, 5)
ax6 = fig.add_subplot(3, 4, 6)
ax7 = fig.add_subplot(3, 4, 7)
ax8 = fig.add_subplot(3, 4, 8)
ax9 = fig.add_subplot(3, 4, 9)
ax10 = fig.add_subplot(3, 4, 10)
ax11 = fig.add_subplot(3, 4, 11)

ax_list = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10, ax11]

for i in range(11):
    box_labels = make_confusion_matrix(cf_list[i], 
                      group_names=labels,
                      categories=categories, percent=False, title=kenya_title[i])
    sns.heatmap(cf_list[i], annot=box_labels,fmt="",cmap="RdBu", ax=ax_list[i], cbar_kws={'shrink': .3}, 
                square=True, annot_kws={'fontsize': 18}, xticklabels=categories,yticklabels=categories)
    ax_list[i].set_title(kenya_title[i])
    ax_list[i].set_xlabel("Stunting Predicted")
    ax_list[i].set_ylabel("Actually Stunted")

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.65, 
                    wspace=0.35, 
                    hspace=0.2)
# plt.title(label="Confusion Matrices for Kenya Models", fontsize=30, color="black")
plt.savefig('kenya_cm.png', dpi=600, bbox_inches='tight')
```

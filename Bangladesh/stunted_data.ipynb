```python
import pandas as pd
```

```python
data = pd.read_csv("rsrc.csv")
```

```python
targetVar = ["Participant_Id", "Median mother's height (cm) [EUPATH_0035110]", "Sex [PATO_0000047]", 
            "Mother's age (years) [EUPATH_0035109]", "Cluster study arm [EUPATH_0044124]",
            "Household Food Insecurity Access Scale (HFIAS) score [EUPATH_0011151]", "Birth order [EUPATH_0035114]",
            "Maternal education (years) [EUPATH_0011605]", "Paternal education (years) [EUPATH_0035043]"]
features = data[targetVar].dropna()
features
```

```python
# These are the main assets which can help determine wealth for each individual household
# They are comprised of tech
# The variables are weighted to consider the approximate cost
asset = ["Participant_Id", "Bicycle [ENVO_01000614]", "Black and white television [EUPATH_0022011]", 
        "Color television [EUPATH_0022012]", "Electricity [EUPATH_0021084]", "Refrigerator [ENVO_01000583]", 
        "Motorcycle or scooter [ENVO_01000615]", "Acres [EUPATH_0000026]"]
encodeYN = {"Yes":1, "No":0}
assetWeight = {asset[1]:0.25, asset[2]:1, asset[3]:1.5, asset[4]:1.5, asset[5]:1, asset[6]:2}

assetVar = data[asset].dropna()
assetVar["Assets"] = assetVar["Acres [EUPATH_0000026]"]*2

for i in range(6):
    a = asset[i+1]
    weight = assetWeight[a]
    assetVar[a] = assetVar[a].replace(encodeYN)
    assetVar["Assets"] += assetVar[a]*weight
    assetVar = assetVar.drop(a, axis=1)
    
assetVar = assetVar.drop("Acres [EUPATH_0000026]", axis=1)
assetVar
```

```python
features = pd.merge(features, assetVar, on="Participant_Id").dropna()
features
```

```python
labelVar = ["Participant_Id", "Stunted [EUPATH_0035062]"]
label = data[data["Study timepoint [OBI_0001508]"]==2]
label = label[labelVar].dropna()
label["Stunted [EUPATH_0035062]"] = label["Stunted [EUPATH_0035062]"].replace(encodeYN)
label
```

```python
stunted_data = pd.merge(label, features, on="Participant_Id").dropna()
stunted_data
```

```python
encodeSex = {"Female":1, "Male":0}
stunted_data["Sex [PATO_0000047]"] = stunted_data["Sex [PATO_0000047]"].replace(encodeSex)
stunted_data = pd.get_dummies(stunted_data, columns=["Cluster study arm [EUPATH_0044124]"], drop_first=True)
```
